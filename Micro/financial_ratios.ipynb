```python
#import libraries
import os
import requests
import pandas as pd
import numpy as np
import plotly.figure_factory as ff
```

```python
#import environment variable
import alpha_vantage as av
av_api_key = os.getenv("ALPHA_VANTAGE_API_KEY")
```

```python
def get_financial_ratios(ticker):
    base_url1 = 'https://www.alphavantage.co/query?'
    function1 = 'BALANCE_SHEET'

    response1 = requests.get(f'{base_url1}function={function1}&symbol={ticker}&apikey={av_api_key}') 
    overview1 = response1.json()

    balance_overview = overview1['annualReports']
    df_balance = pd.DataFrame(balance_overview)

    base_url2 = 'https://www.alphavantage.co/query?'
    function2 = 'INCOME_STATEMENT'

    response2 = requests.get(f'{base_url2}function={function2}&symbol={ticker}&apikey={av_api_key}') 
    overview2 = response2.json()

    income_overview = overview2['annualReports']
    df_income = pd.DataFrame(income_overview)

    base_url3 = 'https://www.alphavantage.co/query?'
    function3 = 'CASH_FLOW'

    response3 = requests.get(f'{base_url3}function={function3}&symbol={ticker}&apikey={av_api_key}') 
    overview3 = response3.json()

    cash_overview = overview3['annualReports']
    df_cash = pd.DataFrame(cash_overview)

    #Liquidity ratios
    #Current ratio = Current assets / Current liabilities
    #Acid-test ratio = Current assets – Inventories / Current liabilities
    #Cash ratio = Cash and Cash equivalents / Current Liabilities
    #Operating cash flow ratio = Operating cash flow / Current liabilities

    current_ratio1 = float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[0][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[0][1])
    quick_ratio1 = (float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[0][1]) - float(df_balance[['fiscalDateEnding', 'inventory']].iloc[0][1])) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[0][1])
    cash_ratio1 = float(df_balance[['fiscalDateEnding', 'cash']].iloc[0][1])/float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[0][1])
    op_cash_ratio1 = float(df_cash[['fiscalDateEnding', 'operatingCashflow']].iloc[0][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[0][1])

    current_ratio2 = float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[1][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[1][1])
    quick_ratio2 = (float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[1][1]) - float(df_balance[['fiscalDateEnding', 'inventory']].iloc[1][1])) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[0][1])
    cash_ratio2 = float(df_balance[['fiscalDateEnding', 'cash']].iloc[1][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[1][1])
    op_cash_ratio2= float(df_cash[['fiscalDateEnding', 'operatingCashflow']].iloc[1][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[1][1])

    current_ratio3 = float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[2][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[2][1])
    quick_ratio3 = (float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[2][1]) - float(df_balance[['fiscalDateEnding', 'inventory']].iloc[2][1])) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[0][1])
    cash_ratio3 = float(df_balance[['fiscalDateEnding', 'cash']].iloc[2][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[2][1])
    op_cash_ratio3 = float(df_cash[['fiscalDateEnding', 'operatingCashflow']].iloc[2][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[2][1])

    current_ratio4 = float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[3][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[3][1])
    quick_ratio4 = (float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[3][1]) - float(df_balance[['fiscalDateEnding', 'inventory']].iloc[3][1])) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[0][1])
    cash_ratio4 = float(df_balance[['fiscalDateEnding', 'cash']].iloc[3][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[3][1])
    op_cash_ratio4 = float(df_cash[['fiscalDateEnding', 'operatingCashflow']].iloc[3][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[3][1])

    current_ratio5 = float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[4][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[4][1])
    quick_ratio5 = (float(df_balance[['fiscalDateEnding', 'totalCurrentAssets']].iloc[4][1]) - float(df_balance[['fiscalDateEnding', 'inventory']].iloc[4][1])) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[0][1])
    cash_ratio5 = float(df_balance[['fiscalDateEnding', 'cash']].iloc[4][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[4][1])
    op_cash_ratio5 = float(df_cash[['fiscalDateEnding', 'operatingCashflow']].iloc[4][1]) / float(df_balance[['fiscalDateEnding', 'totalCurrentLiabilities']].iloc[4][1])

    #Liquidity ratios combined
    liquid_ratios_FY1 = pd.DataFrame([current_ratio1, quick_ratio1, cash_ratio1, op_cash_ratio1], columns=[df_balance.iloc[0][0]], index = ['Current Ratio', 'Quick Ratio', 'Cash Ratio', 'OCF Ratio'])
    liquid_ratios_FY2 = pd.DataFrame([current_ratio2, quick_ratio2, cash_ratio2, op_cash_ratio2], columns=[df_balance.iloc[1][0]], index = ['Current Ratio', 'Quick Ratio', 'Cash Ratio', 'OCF Ratio'])
    liquid_ratios_FY3 = pd.DataFrame([current_ratio3, quick_ratio3, cash_ratio3, op_cash_ratio3], columns=[df_balance.iloc[2][0]], index = ['Current Ratio', 'Quick Ratio', 'Cash Ratio', 'OCF Ratio'])
    liquid_ratios_FY4 = pd.DataFrame([current_ratio4, quick_ratio4, cash_ratio4, op_cash_ratio4], columns=[df_balance.iloc[3][0]], index = ['Current Ratio', 'Quick Ratio', 'Cash Ratio', 'OCF Ratio'])
    liquid_ratios_FY5 = pd.DataFrame([current_ratio5, quick_ratio5, cash_ratio5, op_cash_ratio5], columns=[df_balance.iloc[4][0]], index = ['Current Ratio', 'Quick Ratio', 'Cash Ratio', 'OCF Ratio'])


    #Leverage ratios
    #Debt ratio = Total liabilities / Total assets
    #Debt to equity ratio = Total liabilities / Shareholder’s equity
    #Interest coverage ratio = Operating income / Interest expenses

    debt_ratio1 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[0][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[0][1])
    debt_equity_ratio1 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[0][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[0][1])
    int_cov_ratio1 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[0][1]) / float(df_income[['fiscalDateEnding', 'interestExpense']].iloc[0][1])

    debt_ratio2 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[1][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[1][1])
    debt_equity_ratio2 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[1][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[1][1])
    int_cov_ratio2 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[1][1]) / float(df_income[['fiscalDateEnding', 'interestExpense']].iloc[1][1])

    debt_ratio3 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[2][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[2][1])
    debt_equity_ratio3 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[2][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[2][1])
    int_cov_ratio3 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[2][1]) / float(df_income[['fiscalDateEnding', 'interestExpense']].iloc[2][1])

    debt_ratio4 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[3][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[3][1])
    debt_equity_ratio4 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[3][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[3][1])
    int_cov_ratio4 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[3][1]) / float(df_income[['fiscalDateEnding', 'interestExpense']].iloc[3][1])

    debt_ratio5 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[4][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[4][1])
    debt_equity_ratio5 = float(df_balance[['fiscalDateEnding', 'totalLiabilities']].iloc[4][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[4][1])
    int_cov_ratio5 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[4][1]) / float(df_income[['fiscalDateEnding', 'interestExpense']].iloc[4][1])  


    #Leverage ratios combined
    leverage_ratios_FY1 = pd.DataFrame([debt_ratio1, debt_equity_ratio1, int_cov_ratio1], columns=[df_balance.iloc[0][0]], index = ['Debt Ratio', 'D/E Ratio', 'ICR Ratio'])
    leverage_ratios_FY2 = pd.DataFrame([debt_ratio2, debt_equity_ratio2, int_cov_ratio2], columns=[df_balance.iloc[1][0]], index = ['Debt Ratio', 'D/E Ratio', 'ICR Ratio'])
    leverage_ratios_FY3 = pd.DataFrame([debt_ratio3, debt_equity_ratio3, int_cov_ratio3], columns=[df_balance.iloc[2][0]], index = ['Debt Ratio', 'D/E Ratio', 'ICR Ratio'])
    leverage_ratios_FY4 = pd.DataFrame([debt_ratio4, debt_equity_ratio4, int_cov_ratio4], columns=[df_balance.iloc[3][0]], index = ['Debt Ratio', 'D/E Ratio', 'ICR Ratio'])
    leverage_ratios_FY5 = pd.DataFrame([debt_ratio5, debt_equity_ratio5, int_cov_ratio5], columns=[df_balance.iloc[4][0]], index = ['Debt Ratio', 'D/E Ratio', 'ICR Ratio'])


    #Efficiency Ratios
    #Asset turnover ratio = Net sales / Total assets
    #Inventory turnover ratio = Cost of goods sold / Average inventory
    #Days sales in inventory ratio = 365 days / Inventory turnover ratio

    asset_turn_ratio1 = float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[0][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[0][1])
    inv_turn_ratio1 = float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[0][1]) / float(df_balance[['fiscalDateEnding', 'inventory']].iloc[0][1])
    day_sale_ratio1 = (float(df_balance[['fiscalDateEnding', 'inventory']].iloc[0][1]) / float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[0][1])) * 365

    asset_turn_ratio2 = float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[1][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[1][1])
    inv_turn_ratio2 = float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[1][1]) / float(df_balance[['fiscalDateEnding', 'inventory']].iloc[1][1])
    day_sale_ratio2 = (float(df_balance[['fiscalDateEnding', 'inventory']].iloc[1][1]) / float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[1][1])) * 365

    asset_turn_ratio3 = float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[2][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[2][1])
    inv_turn_ratio3 = float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[2][1]) / float(df_balance[['fiscalDateEnding', 'inventory']].iloc[2][1])
    day_sale_ratio3 = (float(df_balance[['fiscalDateEnding', 'inventory']].iloc[2][1]) / float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[2][1])) * 365

    asset_turn_ratio4 = float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[3][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[3][1])
    inv_turn_ratio4 = float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[3][1]) / float(df_balance[['fiscalDateEnding', 'inventory']].iloc[3][1])
    day_sale_ratio4 = (float(df_balance[['fiscalDateEnding', 'inventory']].iloc[3][1]) / float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[3][1])) * 365

    asset_turn_ratio5 = float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[4][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[4][1])
    inv_turn_ratio5 = float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[4][1]) / float(df_balance[['fiscalDateEnding', 'inventory']].iloc[4][1])
    day_sale_ratio5 = (float(df_balance[['fiscalDateEnding', 'inventory']].iloc[4][1]) / float(df_income[['fiscalDateEnding', 'costOfRevenue']].iloc[4][1])) * 365

    #Efficiency Ratios combined
    efficiency_ratios_FY1 = pd.DataFrame([asset_turn_ratio1, inv_turn_ratio1, day_sale_ratio1], columns=[df_balance.iloc[0][0]], index = ['ATO Ratio', 'ITO Ratio', 'DSI Ratio'])
    efficiency_ratios_FY2 = pd.DataFrame([asset_turn_ratio2, inv_turn_ratio2, day_sale_ratio2], columns=[df_balance.iloc[1][0]], index = ['ATO Ratio', 'ITO Ratio', 'DSI Ratio'])
    efficiency_ratios_FY3 = pd.DataFrame([asset_turn_ratio3, inv_turn_ratio3, day_sale_ratio3], columns=[df_balance.iloc[2][0]], index = ['ATO Ratio', 'ITO Ratio', 'DSI Ratio'])
    efficiency_ratios_FY4 = pd.DataFrame([asset_turn_ratio4, inv_turn_ratio4, day_sale_ratio4], columns=[df_balance.iloc[3][0]], index = ['ATO Ratio', 'ITO Ratio', 'DSI Ratio'])
    efficiency_ratios_FY5 = pd.DataFrame([asset_turn_ratio5, inv_turn_ratio5, day_sale_ratio5], columns=[df_balance.iloc[4][0]], index = ['ATO Ratio', 'ITO Ratio', 'DSI Ratio'])


    #Profitability Ratios
    #Gross margin ratio = Gross profit / Net sales
    #Operating margin ratio = Operating income / Net sales
    #Return on assets ratio = Net income / Total assets
    #Return on equity ratio = Net income / Shareholder’s equity

    gross_margin_ratio1 = float(df_income[['fiscalDateEnding', 'grossProfit']].iloc[0][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[0][1])
    op_margin_ratio1 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[0][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[0][1])
    roa1 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[0][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[0][1])) 
    roe1 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[0][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[0][1]))

    gross_margin_ratio2 = float(df_income[['fiscalDateEnding', 'grossProfit']].iloc[1][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[1][1])
    op_margin_ratio2 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[1][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[1][1])
    roa2 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[1][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[1][1])) 
    roe2 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[1][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[1][1]))

    gross_margin_ratio3 = float(df_income[['fiscalDateEnding', 'grossProfit']].iloc[2][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[2][1])
    op_margin_ratio3 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[2][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[2][1])
    roa3 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[2][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[2][1])) 
    roe3 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[2][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[2][1]))

    gross_margin_ratio4 = float(df_income[['fiscalDateEnding', 'grossProfit']].iloc[3][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[3][1])
    op_margin_ratio4 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[3][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[3][1])
    roa4 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[3][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[3][1])) 
    roe4 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[3][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[3][1]))

    gross_margin_ratio5 = float(df_income[['fiscalDateEnding', 'grossProfit']].iloc[4][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[4][1])
    op_margin_ratio5 = float(df_income[['fiscalDateEnding', 'ebit']].iloc[4][1]) / float(df_income[['fiscalDateEnding', 'totalRevenue']].iloc[4][1])
    roa5 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[4][1]) / float(df_balance[['fiscalDateEnding', 'totalAssets']].iloc[4][1])) 
    roe5 = (float(df_income[['fiscalDateEnding', 'netIncome']].iloc[4][1]) / float(df_balance[['fiscalDateEnding', 'totalShareholderEquity']].iloc[4][1]))

    #Profitability Ratios combined
    profit_ratios_FY1 = pd.DataFrame([gross_margin_ratio1, op_margin_ratio1, roa1, roe1], columns=[df_balance.iloc[0][0]], index = ['GM Ratio', 'OM Ratio', 'ROA Ratio', 'ROE Ratio'])
    profit_ratios_FY2 = pd.DataFrame([gross_margin_ratio2, op_margin_ratio2, roa2, roe2], columns=[df_balance.iloc[1][0]], index = ['GM Ratio', 'OM Ratio', 'ROA Ratio', 'ROE Ratio'])
    profit_ratios_FY3 = pd.DataFrame([gross_margin_ratio3, op_margin_ratio3, roa3, roe3], columns=[df_balance.iloc[2][0]], index = ['GM Ratio', 'OM Ratio', 'ROA Ratio', 'ROE Ratio'])
    profit_ratios_FY4 = pd.DataFrame([gross_margin_ratio4, op_margin_ratio4, roa4, roe4], columns=[df_balance.iloc[3][0]], index = ['GM Ratio', 'OM Ratio', 'ROA Ratio', 'ROE Ratio'])
    profit_ratios_FY5 = pd.DataFrame([gross_margin_ratio5, op_margin_ratio5, roa5, roe5], columns=[df_balance.iloc[4][0]], index = ['GM Ratio', 'OM Ratio', 'ROA Ratio', 'ROE Ratio'])

    #Each ratio category commbined
    combined1 = pd.concat([liquid_ratios_FY1, leverage_ratios_FY1, efficiency_ratios_FY1, profit_ratios_FY1])
    combined2 = pd.concat([liquid_ratios_FY2, leverage_ratios_FY2, efficiency_ratios_FY2, profit_ratios_FY2])
    combined3 = pd.concat([liquid_ratios_FY3, leverage_ratios_FY3, efficiency_ratios_FY3, profit_ratios_FY3])
    combined4 = pd.concat([liquid_ratios_FY4, leverage_ratios_FY4, efficiency_ratios_FY4, profit_ratios_FY4])
    combined5 = pd.concat([liquid_ratios_FY5, leverage_ratios_FY5, efficiency_ratios_FY5, profit_ratios_FY5])

    #Ratios combined for each Fiscal Year
    total = pd.concat([round(combined1, 2), round(combined2, 2), round(combined3, 2), round(combined4, 2), round(combined5,2)], axis=1)

#Creating the table
    financial_ratios = ff.create_table(total, index=True)

    for i in range(len(financial_ratios.layout.annotations)):
        financial_ratios.layout.annotations[i].font.size = 14

    return financial_ratios
```
