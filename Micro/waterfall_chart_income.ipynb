```python
#import libraries
import os
import requests
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

```python
#import  environment variable
import alpha_vantage as av
av_api_key = os.getenv("ALPHA_VANTAGE_API_KEY")
```

```python
def get_income_chart(ticker):
    base_url = 'https://www.alphavantage.co/query?'
    function = 'INCOME_STATEMENT'

    response = requests.get(f'{base_url}function={function}&symbol={ticker}&apikey={av_api_key}') 
    overview = response.json()

    annual_overview = overview['annualReports']
    df = pd.DataFrame(annual_overview)
    df_income_statement = df.replace({'None': 0})

    rev_0 = float(df_income_statement[['fiscalDateEnding', 'totalRevenue']].iloc[0][1])
    cogs_0 = float(df_income_statement[['fiscalDateEnding', 'costOfRevenue']].iloc[0][1]) * -1
    gross_prof_0 = float(df_income_statement[['fiscalDateEnding', 'grossProfit']].iloc[0][1])
    rd_0 = float(df_income_statement[['fiscalDateEnding', 'researchAndDevelopment']].iloc[0][1]) * -1
    ga_0 = float(df_income_statement[['fiscalDateEnding', 'sellingGeneralAdministrative']].iloc[0][1]) * -1
    opex_0 = float(df_income_statement[['fiscalDateEnding', 'totalOperatingExpense']].iloc[0][1]) * -1
    ebit_0 = float(df_income_statement[['fiscalDateEnding', 'operatingIncome']].iloc[0][1]) 
    other_inc_0 = float(df_income_statement[['fiscalDateEnding', 'otherNonOperatingIncome']].iloc[0][1]) 
    ebt_0 = float(df_income_statement[['fiscalDateEnding', 'incomeBeforeTax']].iloc[0][1]) 
    inc_tax_0 = float(df_income_statement[['fiscalDateEnding', 'taxProvision']].iloc[0][1]) * -1
    net_inc_0 = float(df_income_statement[['fiscalDateEnding', 'netIncome']].iloc[0][1])  

    rev_1 = float(df_income_statement[['fiscalDateEnding', 'totalRevenue']].iloc[1][1])
    cogs_1 = float(df_income_statement[['fiscalDateEnding', 'costOfRevenue']].iloc[1][1]) * -1
    gross_prof_1 = float(df_income_statement[['fiscalDateEnding', 'grossProfit']].iloc[1][1])
    rd_1 = float(df_income_statement[['fiscalDateEnding', 'researchAndDevelopment']].iloc[1][1]) * -1
    ga_1 = float(df_income_statement[['fiscalDateEnding', 'sellingGeneralAdministrative']].iloc[1][1]) * -1
    opex_1 = float(df_income_statement[['fiscalDateEnding', 'totalOperatingExpense']].iloc[1][1]) * -1
    ebit_1 = float(df_income_statement[['fiscalDateEnding', 'operatingIncome']].iloc[1][1]) 
    other_inc_1 = float(df_income_statement[['fiscalDateEnding', 'otherNonOperatingIncome']].iloc[1][1]) 
    ebt_1 = float(df_income_statement[['fiscalDateEnding', 'incomeBeforeTax']].iloc[1][1]) 
    inc_tax_1 = float(df_income_statement[['fiscalDateEnding', 'taxProvision']].iloc[1][1]) * -1
    net_inc_1 = float(df_income_statement[['fiscalDateEnding', 'netIncome']].iloc[1][1])  

    date0 = df_income_statement['fiscalDateEnding'].iloc[0]
    date1 = df_income_statement['fiscalDateEnding'].iloc[1]
   
    income_chart1 = go.Figure(go.Waterfall(
        name = "20", orientation = "v",
        measure = ["relative", "relative", "total", "relative", "relative", "total", "total", "relative", 'total', 'relative', 'total'],
        x = ["Revenue", "COGS", "Gross Profit", "R&D", "G&A", "Opex", "EBIT", "Other Income/Expense", "EBT", "Income Tax Exp", "Net Income"],
        textposition = "outside",
        #text = [rev_0/1000, cogs_0/1000, gross_prof_0/1000, rd_0/1000, ga_0/1000, opex_0/1000, ebit_0/1000, other_inc_0/1000, ebt_0/1000, inc_tax_0/1000, net_inc_0/1000],
        y = [rev_0, cogs_0, gross_prof_0,rd_0, ga_0, opex_0, ebit_0, other_inc_0, ebt_0, inc_tax_0, net_inc_0],
        connector = {"line":{"color":"rgb(63, 63, 63)"}},
    ))

    income_chart1.update_layout(
            title = f"Profit and Loss Statement for {ticker} for FY Ending {date0}",
            showlegend = True
    )


    income_chart2 = go.Figure(go.Waterfall(
        name = "20", orientation = "v",
        measure = ["relative", "relative", "total", "relative", "relative", "total", "total", "relative", 'total', 'relative', 'total'],
        x = ["Revenue", "COGS", "Gross Profit", "R&D", "G&A", "Opex", "EBIT", "Other Income/Expense", "EBT", "Income Tax Exp", "Net Income"],
        textposition = "outside",
        #text = [rev_1/1000, cogs_1/1000, gross_prof_1/1000, rd_1/1000, ga_1/1000, opex_1/1000, ebit_1/1000, other_inc_1/1000, ebt_1/1000, inc_tax_1/1000, net_inc_1/1000],
        y = [rev_1, cogs_1, gross_prof_1, rd_1, ga_1, opex_1, ebit_1, other_inc_1, ebt_1, inc_tax_1, net_inc_1],
        connector = {"line":{"color":"rgb(63, 63, 63)"}},
    ))

    income_chart2.update_layout(
            title = f"Profit and Loss Statement for {ticker} for FY Ending {date1}",
            showlegend = True
    )
    
    return income_chart1, income_chart2
```
